```julia
using NBInclude
@nbinclude("./Project/Setup.ipynb");
using Plots; pyplot()
```

```julia
#Load 4-allele model
@nbinclude("DoubleDrive_Simulator_4allele.ipynb");
```

```julia
#Assign plotting parameters
PyPlot.rc("font", family="Arial")
PyPlot.rc("xtick", labelsize=12)
PyPlot.rc("ytick", labelsize=12)

blue = "#0155FF"
light_blue = "#78A0FF"
red = "#FF011A"
light_red = "#FF7689"
col1 = "#BA03FF" #purple
col2 = "#FF8509" #orange
col3 = "#FF6CE8" #pink
col4 = "#FFBF09" #yellow
col5 = "#00DF97" #green
```

```julia
constructs = [
["Construct_locA-neutral_locB-HS_rec_homA-constitutive_homB-conditional_XS-none",col1],
["Construct_locA-neutral_locB-HS_rec_homA-conditional_homB-conditional_XS-none",col2],
["Construct_locA-neutral_locB-HS_rec_homA-constitutive_homB-conditional_XS-constitutive_B",col3],
["Construct_locA-neutral_locB-HS_rec_homA-conditional_homB-conditional_XS-constitutive_B",col4],
["Construct_locA-neutral_locB-HI_homA-conditional_homB-constitutive_XS-constitutive_A",col5],
];
```

```julia
#PLOT
f1, axes = PyPlot.subplots(3,5,figsize=(12,6),frameon=false,sharex=true,sharey=true)
t=200
n=0
for c in constructs

    #unpack
    construct_label = c[1]
    LOCUS_A,
    LOCUS_B,
    set_A_homing_dependency,
    set_B_homing_dependency,
    set_XS_dependency = unpack_construct_label(construct_label)

    #Assign starting parameters.
    params,
    A_homing_dependency,
    B_homing_dependency,
    XS_dependency = Assign_baseline_params(LOCUS_A,
                LOCUS_B,
                set_A_homing_dependency,
                set_B_homing_dependency,
                set_XS_dependency);

    A_construct = []
    B_construct = []
    fitness_effect = "adult"
    transgenic_genotype = "ABab"
    release_freq = 0.001
    resistant_allele = "α"
    resistant_freq = 1.0
    Aallele_fitness = "α"

    #SImulate baseline 
    output = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype,
                                release_freq, 
                                resistant_allele, 
                                resistant_freq,
                                t);
    

    #simulate sep release
    output_sep = Simulation_wrapper_separate_release(params,
                            A_construct,
                            B_construct,
                            A_homing_dependency,
                            B_homing_dependency,
                            XS_dependency,
                            Aallele_fitness,
                            fitness_effect,
                            "ABaB",
                            "ABAb",
                            release_freq/2,
                            release_freq/2, 
                            resistant_allele, 
                            resistant_freq,
                            t)
    
    #simulate r=0.01
    params["r"] = 0.01
    output_r = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype,
                                release_freq, 
                                resistant_allele, 
                                resistant_freq,
                                t);

    #simulate insertion into essential gene
    resistant_allele = "o"
    
    #HS
    params["r"] = 0.5
    params["sraf"] = 1.0
    params["sram"] = 1.0
    output_HS = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype,
                                release_freq, 
                                resistant_allele, 
                                resistant_freq,
                                t);
    
    
    #HI
    params["r"] = 0.5
    params["sraf"] = 1.0
    params["hraf"] = 1.0
    params["siraf"] = 1.0
    params["sram"] = 1.0
    params["hram"] = 1.0
    params["siram"] = 1.0
    output_HI = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype,
                                release_freq, 
                                resistant_allele, 
                                resistant_freq,
                                t);
    n=n+1
    ## baseline
    axes[n].plot(0:t,output["allelefreq_adult"].b,c=blue,"-",label = "construct B")
    axes[n].plot(0:t,output["allelefreq_adult"].a,c=red,"-",label = "construct A")
    axes[n].plot(0:t,1 .- output["relative_number_females"].adult,c="black","-", label = "Biting adult females")
    ## r=0.01
    axes[n].plot(0:t,output_r["allelefreq_adult"].b,c=blue,"--",label = "construct B")
    axes[n].plot(0:t,output_r["allelefreq_adult"].a,c=red,"--",label = "construct A")
    axes[n].plot(0:t,1 .- output_r["relative_number_females"].adult,c="black","--", label = "Biting adult females")
    axes[n].set_xlim([0,t])
    axes[n].set_ylim([10^-5,1])
    axes[n].set_yscale("log")

    n=n+1
    ## baseline
    axes[n].plot(0:t,output["allelefreq_adult"].b,c=blue,"-",label = "construct B")
    axes[n].plot(0:t,output["allelefreq_adult"].a,c=red,"-",label = "construct A")
    axes[n].plot(0:t,1 .- output["relative_number_females"].adult,c="black","-", label = "Biting adult females")
    ## HI
    axes[n].plot(0:t,output_HI["allelefreq_adult"].b,c=blue,"--",label = "construct B")
    axes[n].plot(0:t,output_HI["allelefreq_adult"].a,c=red,"--",label = "construct A")
    axes[n].plot(0:t,1 .- output_HI["relative_number_females"].adult,c="black","--", label = "Biting adult females")
    ## HS
    axes[n].plot(0:t,output_HS["allelefreq_adult"].b,c=blue,":",label = "construct B")
    axes[n].plot(0:t,output_HS["allelefreq_adult"].a,c=red,":",label = "construct A")
    axes[n].plot(0:t,1 .- output_HS["relative_number_females"].adult,c="black",":", label = "Biting adult females")
    axes[n].set_xlim([0,t])
    axes[n].set_ylim([10^-5,1])
    axes[n].set_yscale("log")

    n=n+1
    ## baseline
    axes[n].plot(0:t,output["allelefreq_adult"].b,c=blue,"-",label = "construct B")
    axes[n].plot(0:t,output["allelefreq_adult"].a,c=red,"-",label = "construct A")
    axes[n].plot(0:t,1 .- output["relative_number_females"].adult,c="black","-", label = "Biting adult females")
    ## sep release
    axes[n].plot(0:t,output_sep["allelefreq_adult"].b,c=blue,"--",label = "construct B")
    axes[n].plot(0:t,output_sep["allelefreq_adult"].a,c=red,"--",label = "construct A")
    axes[n].plot(0:t,1 .- output_sep["relative_number_females"].adult,c="black","--", label = "Biting adult females")
    axes[n].set_xlim([0,t])
    axes[n].set_ylim([10^-5,1])
    axes[n].set_yscale("log")
    
end

PyPlot.tight_layout()
PyPlot.show()
```
