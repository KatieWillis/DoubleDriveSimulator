### Example simulations of releases in different males

```julia
#Define design 1
LOCUS_A = "neutral"
LOCUS_B = "HS_rec"
set_A_homing_dependency = "constitutive"
set_B_homing_dependency = "conditional"
set_XS_dependency = "none"

#Assign starting parameters.
params,
A_homing_dependency,
B_homing_dependency,
XS_dependency = Assign_baseline_params(LOCUS_A,
            LOCUS_B,
            set_A_homing_dependency,
            set_B_homing_dependency,
            set_XS_dependency);

A_construct = []
B_construct = []
fitness_effect = "adult"
transgenic_genotype = "ABab"
release_freq = 0.001
resistant_freq = 0.01

resistant_allele = "α"
Aallele_fitness = "α"

t = 100
```

```julia
#Same males
output_1a = Simulation_wrapper(params,
                            A_construct,
                            B_construct,
                            A_homing_dependency,
                            B_homing_dependency,
                            XS_dependency,
                            Aallele_fitness,
                            fitness_effect,
                            transgenic_genotype,
                            release_freq, 
                            resistant_allele, 
                            resistant_freq,
                            t)
#separate males
output_1b = Simulation_wrapper_separate_release(params,
                            A_construct,
                            B_construct,
                            A_homing_dependency,
                            B_homing_dependency,
                            XS_dependency,
                            Aallele_fitness,
                            fitness_effect,
                            "ABaB",
                            "ABAb",
                            release_freq/2,
                            release_freq/2, 
                            resistant_allele, 
                            resistant_freq,
                            t);
```

```julia
println(minimum(output_1a["relative_number_females"].adult))
println(minimum(output_1b["relative_number_females"].adult))
```

```julia
#plot
PyPlot.rc("font", family="Arial")
PyPlot.rc("xtick", labelsize=12) 
PyPlot.rc("ytick", labelsize=12)

f1, axes = PyPlot.subplots(2,1,figsize=(5,5),frameon=false,)

output = output_1a
t=100
axes[1].plot(0:t,output["relative_number_females"].adult,c="black","--", label = "Biting adult females")
axes[1].plot(0:t,output["allelefreq_adult"].b,c=blue,"-",label = "construct B")
axes[1].plot(0:t,output["allelefreq_adult"].B,c=blue,"--",label = "WT")
axes[1].plot(0:t,output["allelefreq_adult"].a,c=red,"-",label = "construct A")
axes[1].plot(0:t,output["allelefreq_adult"].α,c=red,"--",label = "resistant")
axes[1].plot(0:t,output["correlation_adult"].Correlation,c="black",":",label = "correlation")
axes[1].set_xlim([0,t])
axes[1].set_ylim([-0.05,1])

output = output_1b
t=100
axes[2].plot(0:t,output["relative_number_females"].adult,c="black","--", label = "Biting adult females")
axes[2].plot(0:t,output["allelefreq_adult"].b,c=blue,"-",label = "construct B")
axes[2].plot(0:t,output["allelefreq_adult"].B,c=blue,"--",label = "WT")
axes[2].plot(0:t,output["allelefreq_adult"].a,c=red,"-",label = "construct A")
axes[2].plot(0:t,output["allelefreq_adult"].α,c=red,"--",label = "resistant")
axes[2].plot(0:t,output["correlation_adult"].Correlation,c="black",":",label = "correlation")
axes[2].set_xlim([0,t])
axes[2].set_ylim([-0.05,1])

PyPlot.tight_layout()
PyPlot.show()

PyPlot.rc("font", family="Arial")
PyPlot.rc("xtick", labelsize=10) 
PyPlot.rc("ytick", labelsize=10)
```

### Time courses of relative female density

```julia
resistant_allele = "o"
Aallele_fitness = "α"

transgenic_genotype = "ABab"
release_freq = 0.001

transgenic_genotype1 = "ABaB"
transgenic_genotype2 = "ABAb"
release_freq1=release_freq/2
release_freq2=release_freq/2
```

```julia
function plot_different_releases(construct_label, construct_colour,resistant_freq,x_upper,y_lower)

    t=x_upper

    #PLOT example simulation - neutral + HI,HS + sep rel 
    f1 = PyPlot.figure(figsize=(4,2.5),frameon=false)
    
    #function to unpack details from label
    LOCUS_A,
    LOCUS_B,
    set_A_homing_dependency,
    set_B_homing_dependency,
    set_XS_dependency = unpack_construct_label(construct_label)

    #Assign starting parameters.
    params,
    A_homing_dependency,
    B_homing_dependency,
    XS_dependency = Assign_baseline_params(LOCUS_A,
                LOCUS_B,
                set_A_homing_dependency,
                set_B_homing_dependency,
                set_XS_dependency);

    A_construct = []
    B_construct = []
    fitness_effect = "adult"

    #SIMULATE Neutral unlinked
    output_N_unlinked = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype,
                                release_freq, 
                                resistant_allele, 
                                resistant_freq,
                                t);
    output_N_unlinked_sep = Simulation_wrapper_separate_release(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype1,
                                transgenic_genotype2,
                                release_freq1,
                                release_freq2, 
                                resistant_allele, 
                                resistant_freq,
                                t)
    #SIMULATE Neutral unlinked
    params["r"] = 0.01
    output_N_linked = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype,
                                release_freq, 
                                resistant_allele, 
                                resistant_freq,
                                t);
    output_N_linked_sep = Simulation_wrapper_separate_release(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                Aallele_fitness,
                                fitness_effect,
                                transgenic_genotype1,
                                transgenic_genotype2,
                                release_freq1,
                                release_freq2, 
                                resistant_allele, 
                                resistant_freq,
                                t)
    
    
    #Simulate the single-construct model - where NHEJ produces non-functional resistance 
    LOCUS_A = "HS_rec"
    LOCUS_B = "neutral"
    set_A_homing_dependency = "constitutive"
    set_B_homing_dependency = "none"
    set_XS_dependency = "none"

    #Assign starting parameters...
    params,
    A_homing_dependency,
    B_homing_dependency,
    XS_dependency = Assign_baseline_params(LOCUS_A,
                LOCUS_B,
                set_A_homing_dependency,
                set_B_homing_dependency,
                set_XS_dependency);
    
    #The standing resistant allele (o) suffers no costs:
    params["sraf"] = 0.0
    params["siraf"] = 0.0

    output_single_construct = Simulation_wrapper(params,
                                A_construct,
                                B_construct,
                                A_homing_dependency,
                                B_homing_dependency,
                                XS_dependency,
                                "o",
                                fitness_effect,
                                "ABaB",
                                release_freq, 
                                "o", 
                                resistant_freq,
                                t);

    PyPlot.plot(0:t,output_N_unlinked["relative_number_females"].adult,c=construct_colour,"-",label="unlinked, cis")
    PyPlot.plot(0:t,output_N_unlinked_sep["relative_number_females"].adult,c=construct_colour,"--",label="unlinked, trans")
    PyPlot.plot(0:t,output_N_linked["relative_number_females"].adult,c=construct_colour,":",label="linked, cis")
    PyPlot.plot(0:t,output_N_linked_sep["relative_number_females"].adult,c=construct_colour,"-.",label="linked, trans")
    PyPlot.plot(0:t,output_single_construct["relative_number_females"].adult,c="black","-",label="S")

    PyPlot.xlim([0,t])
    PyPlot.ylim([y_lower,10])
    PyPlot.yscale("log")

    PyPlot.yticks([10^-10,10^-8,10^-6,10^-4,10^-2,10^-0],
        ["\$10^{-10}\$","\$10^{-8}\$","\$10^{-6}\$","\$10^{-4}\$","\$10^{-2}\$","1.0"],size=12)
    PyPlot.xticks(size=12)
    PyPlot.tight_layout()
    
    PyPlot.show()
end
```

```julia
#Define constructs to loop through and unpack details 
constructs = [
["Construct_locA-neutral_locB-HS_rec_homA-constitutive_homB-conditional_XS-none",col1],
["Construct_locA-neutral_locB-HS_rec_homA-conditional_homB-conditional_XS-none",col2],
["Construct_locA-neutral_locB-HS_rec_homA-constitutive_homB-conditional_XS-constitutive_B",col3],
["Construct_locA-neutral_locB-HS_rec_homA-conditional_homB-conditional_XS-constitutive_B",col4],
["Construct_locA-neutral_locB-HI_homA-conditional_homB-constitutive_XS-constitutive_A",col5],
];

x_upper_vec = [100, 100, 120, 120, 300]
y_lower_vec = [10^-10,10^-10,10^-12,10^-12,10^-10]
res_freq_vec = [0.01,0.01,0.2,0.2,0.5]

for i in 1:5
    plot_different_releases(constructs[i][1], constructs[i][2],res_freq_vec[i],x_upper_vec[i],y_lower_vec[i])
end
```
